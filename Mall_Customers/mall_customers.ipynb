### Find Behavioral Similarities


```python
data_new[:10]
```

```python
header
```

```python
avg = np.mean(data_new, axis=0)
```

```python
std_dev = np.std(data_new, axis=0)
```

```python
data_norm = (data_new - avg) / std_dev
data_norm
```

```python
data_norm = data_norm[:, 2:]
```

```python
data_norm.shape
```

- https://www.geeksforgeeks.org/calculate-the-euclidean-distance-using-numpy/
- https://stackoverflow.com/questions/1401712/how-can-the-euclidean-distance-be-calculated-with-numpy
- https://www.kdnuggets.com/2023/03/distance-metrics-euclidean-manhattan-minkowski-oh.html

```python
highest_idx = np.argmax(data_norm[:,-1]) 
highest_spender = data_norm[highest_idx]
highest_spender
```

```python
distances = np.linalg.norm(data_norm - highest_spender, axis=1)
distances.shape
```

```python
distances
```

```python
similar_customers_idx = distances <= 1.0
similar_customers = data_new[similar_customers_idx]
similar_customers
```

Younger female customer has similar high spending score.
