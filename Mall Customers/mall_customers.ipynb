### Load Data Using Numpy genfromtxt

```python
data = np.genfromtxt('Mall_Customers.csv', dtype=['i8','U6','i8','i8','i8'], names=['id', 'Gender', 'Age', 'AnnualIncome','Spending'], delimiter=',', skip_header=1)
data
```

**Note:**

- **Use U6 instead of S6**

**Reference**
- https://numpy.org/doc/stable/reference/generated/numpy.genfromtxt.html
- https://stackoverflow.com/questions/33655641/b-character-added-when-using-numpy-loadtxt

```python
# Filter gender with male
# Will use the list to double check when doing conversion
data['Gender'] == 'Male'
```

**We will be converting Male to 1 and Female to 0.**

```python
# Try np.where without actual assignment. Check if conversion is ok.
np.where(data['Gender'] == 'Male', 1, 0)
```

```python
# Perform actual conversion
data['Gender'] = np.where(data['Gender'] == 'Male', 1, 0)
data
```

The gender columns still remain as string.
